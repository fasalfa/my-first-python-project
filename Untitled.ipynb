```python
plt.figure(figsize=(12, 6))
sns.histplot(df, x='price', y='size', bins=30, kde=True)
plt.title('Distribution of House Prices in Bangalore')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x='total_sqft', y='price', data=df)
plt.title('Price vs. Total Square Feet Area')
plt.xlabel('Total Square Feet')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='location', y='price', data=df.groupby('location')['price'].mean().reset_index().sort_values(by='price', ascending=False).head(10))
plt.title('Top 10 Locations by Average Price')
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='price', y='size', data=df)
plt.title('Price Distribution by Number of Bedrooms (BHK)')
plt.xlabel('price')
plt.ylabel('bhk')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le_area_type = LabelEncoder()
df['area_type_encoded'] = le_area_type.fit_transform(df['area_type'])

le_availability = LabelEncoder()
df['availability_encoded'] = le_availability.fit_transform(df['availability'])
```

```python
df
```

```python
df = df.drop(['area_type', 'availability', 'location', 'size'], axis=1)
```

```python
df = df.drop(['society'], axis=1)
```

```python
df
```

```python
def convert_range_to_numeric(value):
    try:
        # Extract numeric values from the range
        numeric_values = [float(val) for val in re.findall(r'\d+', str(value))]
        # Take the average of the range
        return sum(numeric_values) / len(numeric_values)
    except (ValueError, TypeError):
        return None

# Apply the function to 'total_sqft' column
df['total_sqft'] = df['total_sqft'].apply(convert_range_to_numeric)

# Drop rows with NaN values in 'total_sqft'
df = df.dropna(subset=['total_sqft'])

# Now, calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
